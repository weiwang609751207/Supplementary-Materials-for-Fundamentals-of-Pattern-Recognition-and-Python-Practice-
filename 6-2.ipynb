```python
# %load 6-2.py
import numpy as np

def bayesian_decision_theory():
    """
    卤麓脪露脣鹿戮枚虏脽拢潞麓脫脧脠脩茅脰陋脢露碌陆潞贸脩茅戮枚虏脽
    """
    # 鲁隆戮掳拢潞脪陆脕脝脮茂露脧 - 脜脨露脧脢脟路帽禄录脫脨脛鲁脰脰录虏虏隆
    
    # 脧脠脩茅赂脜脗脢拢潞脠脣脠潞脰脨赂脙录虏虏隆碌脛禄录虏隆脗脢
    P_disease = 0.01      # 1%碌脛脠脣驴脷禄录虏隆
    P_healthy = 0.99      # 99%碌脛脠脣驴脷陆隆驴碌
    
    # 脣脝脠禄露脠拢潞录矛虏芒碌脛脳录脠路脨脭
    # P(脩么脨脭|禄录虏隆) = 0.95, P(脩么脨脭|陆隆驴碌) = 0.05
    P_positive_given_disease = 0.95
    P_positive_given_healthy = 0.05
    
    # 录脵脡猫脛鲁脠脣录矛虏芒陆谩鹿没脦陋脩么脨脭拢卢脟贸禄录虏隆碌脛潞贸脩茅赂脜脗脢
    # P(禄录虏隆|脩么脨脭) = P(脩么脨脭|禄录虏隆) * P(禄录虏隆) / P(脩么脨脭)
    P_positive = (P_positive_given_disease * P_disease + 
                  P_positive_given_healthy * P_healthy)
    
    P_disease_given_positive = (P_positive_given_disease * P_disease) / P_positive
    
    print(f"脧脠脩茅禄录虏隆赂脜脗脢: {P_disease:.3f}")
    print(f"录矛虏芒脩么脨脭潞贸禄录虏隆赂脜脗脢: {P_disease_given_positive:.3f}")
    
    return P_disease_given_positive
def risk_based_decision():
    """
    禄霉脫脷路莽脧脮碌脛戮枚虏脽拢潞驴录脗脟虏禄脥卢麓铆脦贸碌脛麓煤录脹
    """
    # 脌卢禄酶脫脢录镁路脰脌脿鲁隆戮掳
    
    # 脧脠脩茅赂脜脗脢
    P_spam = 0.3      # 30%脢脟脌卢禄酶脫脢录镁
    P_ham = 0.7       # 70%脢脟脮媒鲁拢脫脢录镁
    
    # 路脰脌脿脝梅碌脛赂脜脗脢脢盲鲁枚
    P_pred_spam = 0.4  # 路脰脌脿脝梅脠脧脦陋40%驴脡脛脺脢脟脌卢禄酶脫脢录镁
    
    # 路莽脧脮戮脴脮贸拢潞虏禄脥卢戮枚虏脽碌脛麓煤录脹
    # 脨脨拢潞脮忙脢碌脌脿卤冒拢卢脕脨拢潞脭陇虏芒脌脿卤冒
    risk_matrix = np.array([
        [0,  10],   # 掳脩脮媒鲁拢脫脢录镁脜脨脦陋脌卢禄酶脫脢录镁拢潞麓煤录脹10拢篓驴脡脛脺麓铆鹿媒脰脴脪陋脫脢录镁拢漏
        [1,  0]     # 掳脩脌卢禄酶脫脢录镁脜脨脦陋脮媒鲁拢脫脢录镁拢潞麓煤录脹1拢篓脰禄脢脟脫脨碌茫路鲁拢漏
    ])
    
    # 录脝脣茫脕陆脰脰戮枚虏脽碌脛脝脷脥没路莽脧脮
    risk_classify_as_ham = (P_pred_spam * risk_matrix[1, 0] + 
                           (1 - P_pred_spam) * risk_matrix[0, 0])
    
    risk_classify_as_spam = (P_pred_spam * risk_matrix[1, 1] + 
                            (1 - P_pred_spam) * risk_matrix[0, 1])
    
    # 脩隆脭帽路莽脧脮陆脧脨隆碌脛戮枚虏脽
    if risk_classify_as_ham < risk_classify_as_spam:
        decision = "姝ｅ父閭欢"
        risk = risk_classify_as_ham
    else:
        decision = "鍨冨溇閭欢" 
        risk = risk_classify_as_spam
        
    print(f"鍐崇瓥: {decision}, 鏈熸湜椋庨櫓: {risk:.2f}")
    return decision, risk
# 脭脣脨脨脢戮脌媒
result = risk_based_decision()
```
