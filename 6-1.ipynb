```python
# %load 6-1.py
import numpy as np

def bayesian_decision_theory():
    """
    ±´Ò¶Ë¹¾ö²ß£º´ÓÏÈÑéÖªÊ¶µ½ºóÑé¾ö²ß
    """
    # ³¡¾°£ºÒ½ÁÆÕï¶Ï - ÅÐ¶ÏÊÇ·ñ»¼ÓÐÄ³ÖÖ¼²²¡
    
    # ÏÈÑé¸ÅÂÊ£ºÈËÈºÖÐ¸Ã¼²²¡µÄ»¼²¡ÂÊ
    P_disease = 0.01      # 1%µÄÈË¿Ú»¼²¡
    P_healthy = 0.99      # 99%µÄÈË¿Ú½¡¿µ
    
    # ËÆÈ»¶È£º¼ì²âµÄ×¼È·ÐÔ
    # P(ÑôÐÔ|»¼²¡) = 0.95, P(ÑôÐÔ|½¡¿µ) = 0.05
    P_positive_given_disease = 0.95
    P_positive_given_healthy = 0.05
    
    # ¼ÙÉèÄ³ÈË¼ì²â½á¹ûÎªÑôÐÔ£¬Çó»¼²¡µÄºóÑé¸ÅÂÊ
    # P(»¼²¡|ÑôÐÔ) = P(ÑôÐÔ|»¼²¡) * P(»¼²¡) / P(ÑôÐÔ)
    P_positive = (P_positive_given_disease * P_disease + 
                  P_positive_given_healthy * P_healthy)
    
    P_disease_given_positive = (P_positive_given_disease * P_disease) / P_positive
    
    print(f"先验患病概率: {P_disease:.3f}")
    print(f"检测阳性后患病概率: {P_disease_given_positive:.3f}")
    
    return P_disease_given_positive
# ÔËÐÐÊ¾Àý
result = bayesian_decision_theory()
```
