```python
# %load 6-4.py
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置默认字体为黑体
plt.rcParams['axes.unicode_minus'] = False
from sklearn.datasets import make_circles
from sklearn.svm import SVC
def svm_kernel_trick_demo():
    """
    SVMºË¼¼ÇÉÑÝÊ¾£ºÊ¹ÓÃ¸ßË¹ºË´¦Àí·ÇÏßÐÔ¿É·ÖÊý¾Ý
    """
    # Éú³É·ÇÏßÐÔ¿É·ÖÊý¾Ý£¨»·ÐÎÊý¾Ý£©
    X, y = make_circles(n_samples=100, factor=0.5, noise=0.05, random_state=42)
    # ÑµÁ·Ê¹ÓÃ¸ßË¹ºËµÄSVM
    svm_rbf = SVC(kernel='rbf', gamma=10)
    svm_rbf.fit(X, y)
    # »æÖÆ¾ö²ß±ß½ç
    plt.figure(figsize=(10, 6))
    
    # ´´½¨Íø¸ñµã
    x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
    y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200),
                         np.linspace(y_min, y_max, 200))
    
    # Ô¤²âÍø¸ñµãµÄÀà±ð
    Z = svm_rbf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    # »æÖÆ¾ö²ß±ß½çºÍÊý¾Ýµã
    plt.contourf(xx, yy, Z, alpha=0.4)
    plt.scatter(X[:, 0], X[:, 1], c=y, s=20, edgecolors='k')
    
    plt.xlabel('特征1')
    plt.ylabel('特征2')
    plt.title('SVM核技巧：高斯核处理非线性可分数据')
    plt.show()
    return svm_rbf
# ÔËÐÐÑÝÊ¾
svm_rbf_model = svm_kernel_trick_demo()
```
