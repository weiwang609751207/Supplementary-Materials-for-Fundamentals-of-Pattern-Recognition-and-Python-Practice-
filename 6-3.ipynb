```python
# %load 6-3.py
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置默认字体为黑体
plt.rcParams['axes.unicode_minus'] = False
import numpy as np
from sklearn.svm import SVC
from sklearn.datasets import make_blobs

def svm_max_margin_demo():
    """
    SVM×î´ó¼ä¸ôÔ­ÀíÑÝÊ¾
    """
    # Éú³ÉÏßÐÔ¿É·ÖÊý¾Ý
    X, y = make_blobs(n_samples=50, centers=2, 
                      random_state=42, cluster_std=1.0)
    
    # ÑµÁ·ÏßÐÔSVM
    svm_linear = SVC(kernel='linear', C=1000)
    svm_linear.fit(X, y)
    
    # »ñÈ¡¾ö²ß±ß½ç
    w = svm_linear.coef_[0]
    b = svm_linear.intercept_[0]
    
    # »æÖÆ½á¹û
    plt.figure(figsize=(10, 6))
    
    # »æÖÆÊý¾Ýµã
    plt.scatter(X[y==0, 0], X[y==0, 1], c='blue', label='类别0', alpha=0.7)
    plt.scatter(X[y==1, 0], X[y==1, 1], c='red', label='类别1', alpha=0.7)
    
    # »æÖÆ¾ö²ß±ß½ç
    x_plot = np.linspace(X[:, 0].min()-1, X[:, 0].max()+1, 100)
    y_plot = -(w[0] * x_plot + b) / w[1]
    plt.plot(x_plot, y_plot, 'k-', label='决策边界', linewidth=2)
    
    # »æÖÆÖ§³ÖÏòÁ¿
    plt.scatter(svm_linear.support_vectors_[:, 0], 
                svm_linear.support_vectors_[:, 1], 
                s=200, facecolors='none', edgecolors='black', 
                label='支持向量')
    
    plt.xlabel('特征1')
    plt.ylabel('特征2')
    plt.title('SVM最大间隔原理')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()
    
    print(f"支持向量数量: {len(svm_linear.support_vectors_)}")
    return svm_linear

# ÔËÐÐÑÝÊ¾
svm_model = svm_max_margin_demo()
```
