```python
# %load 9-2.py
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置默认字体为黑体
plt.rcParams['axes.unicode_minus'] = False
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
import warnings
warnings.filterwarnings('ignore')

# 设置中文字体和图形样式
#plt.rcParams['font.sans-serif'] = ['SimHei']
#plt.rcParams['axes.unicode_minus'] = False
sns.set_style("whitegrid")

#加载数据（这里使用一个模拟的电商数据集）
#  在实际应用中，你可以从CSV文件加载：
# df = pd.read_csv('customer_data.csv')
np.random.seed(42)

# 创建模拟数据
n_customers = 500
data = {
    'CustomerID': range(1, n_customers + 1),
    'Age': np.random.randint(18, 70, n_customers),
    'AnnualIncome': np.random.normal(50000, 20000, n_customers),
    'SpendingScore': np.random.normal(50, 25, n_customers),
    'TotalPurchases': np.random.poisson(20, n_customers),
    'AvgTransactionValue': np.random.normal(100, 30, n_customers),
    'DaysSinceLastPurchase': np.random.exponential(30, n_customers)
}

df = pd.DataFrame(data)
df['AnnualIncome'] = np.abs(df['AnnualIncome'])  # 确保收入为正数
df['AvgTransactionValue'] = np.abs(df['AvgTransactionValue'])

print("数据概览:")
print(df.head())
print(f"\n数据集形状:{df.shape}")
print("\数据基本信息:")
print(df.info())
print("\n描述性统计:")
print(df.describe())

#绘制特征分布图
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('特征分布分析', fontsize=16)

features = ['Age', 'AnnualIncome', 'SpendingScore', 'TotalPurchases', 'AvgTransactionValue', 'DaysSinceLastPurchase']
colors = ['skyblue', 'lightcoral', 'lightgreen', 'gold', 'plum', 'orange']

for i, (feature, color) in enumerate(zip(features, colors)):
    ax = axes[i//3, i%3]
    df[feature].hist(bins=30, ax=ax, color=color, alpha=0.7)
    ax.set_title(f'{feature}分布')
    ax.set_xlabel(feature)
    ax.set_ylabel('频数')

plt.tight_layout()
plt.show()

# 绘制收入与消费得分的散点图
plt.figure(figsize=(10, 6))
plt.scatter(df['AnnualIncome'], df['SpendingScore'], alpha=0.6, c='blue')
plt.xlabel('年收入')
plt.ylabel('消费得分')
plt.title('年收入 vs 消费得分')
plt.show()

# 计算相关性矩阵
plt.figure(figsize=(10, 8))
correlation_matrix = df[features].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
            square=True, linewidths=0.5)
plt.title('特征相关性热力图')
plt.tight_layout()
plt.show()

# 选择用于聚类的特征
features_for_clustering = ['AnnualIncome', 'SpendingScore', 'TotalPurchases', 'AvgTransactionValue']
X = df[features_for_clustering]

print("聚类特征统计:")
print(X.describe())

# 数据标准化
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 创建标准化后的DataFrame
X_scaled_df = pd.DataFrame(X_scaled, columns=features_for_clustering)
print("\n标准化后的数据统计:")
print(X_scaled_df.describe())

# 使用肘部法则确定最佳K值
inertia = []
silhouette_scores = []
k_range = range(2, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
    
    if k > 1:   # 轮廓系数需要至少2个簇
        silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))

#绘制肘部法则图
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.plot(k_range, inertia, 'bo-', linewidth=2, markersize=8)
plt.xlabel('聚类数量 K')
plt.ylabel('簇内平方和 (Inertia)')
plt.title('肘部法则 - 确定最佳K值')
plt.grid(True)

# 绘制轮廓系数图
plt.subplot(1, 2, 2)
plt.plot(range(2, 11), silhouette_scores, 'ro-', linewidth=2, markersize=8)
plt.xlabel('聚类数量 K')
plt.ylabel('轮廓系数')
plt.title('轮廓系数 - 确定最佳K值')
plt.grid(True)

plt.tight_layout()
plt.show()

# 轮廓系数 - 确定最佳K值
best_k = np.argmax(silhouette_scores) + 2  # +2 ÒòÎª´ÓK=2¿ªÊ¼
print(f"基于轮廓系数的最佳聚类数量:{best_k}")
print(f"最佳轮廓系数: {silhouette_scores[best_k-2]:.4f}")

# 使用最佳K值进行聚类
kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_scaled)

# 将聚类结果添加到原始数据
df['Cluster'] = clusters
df['Cluster'] = df['Cluster'].astype('category')

print("各簇样本数量:")
print(df['Cluster'].value_counts().sort_index())

# 分析每个簇的特征
cluster_profile = df.groupby('Cluster')[features_for_clustering].mean()
cluster_profile['Count'] = df.groupby('Cluster').size()

print("各簇特征均值:")
print(cluster_profile)

# 可视化簇特征对比
plt.figure(figsize=(15, 10))
for i, feature in enumerate(features_for_clustering, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x='Cluster', y=feature, data=df)
    plt.title(f'{feature} ·Ö²¼ by Cluster')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# 使用PCA进行降维可视化（因为特征维度>2）
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# 创建包含PCA结果的DataFrame
pca_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
pca_df['Cluster'] = clusters

# 绘制聚类结果散点图
plt.figure(figsize=(12, 8))
scatter = plt.scatter(pca_df['PC1'], pca_df['PC2'], 
                     c=pca_df['Cluster'], cmap='viridis', alpha=0.7)
plt.colorbar(scatter)
plt.xlabel(f'主成分1 ({pca.explained_variance_ratio_[0]:.2%}方差)')
plt.ylabel(f'主成分2 ({pca.explained_variance_ratio_[1]:.2%}方差)')
plt.title('客户聚类可视化 (PCA降维)')

# 标记簇中心
centers_pca = pca.transform(kmeans.cluster_centers_)
plt.scatter(centers_pca[:, 0], centers_pca[:, 1], c='red', marker='X', s=200, label='簇中心')
plt.legend()

plt.tight_layout()
plt.show()

#  生成详细的簇分析报告
def generate_cluster_report(df, features):
    report = {}
    
    for cluster in sorted(df['Cluster'].unique()):
        cluster_data = df[df['Cluster'] == cluster]
        cluster_size = len(cluster_data)
        
        report[cluster] = {
            'size': cluster_size,
            'percentage': cluster_size / len(df) * 100,
            'characteristics': {}
        }
        
        # 分析每个特征
        for feature in features:
            stats = {
                'mean': cluster_data[feature].mean(),
                'std': cluster_data[feature].std(),
                'median': cluster_data[feature].median()
            }
            report[cluster]['characteristics'][feature] = stats
    
    return report

#生成报告
cluster_report = generate_cluster_report(df, features_for_clustering)

print("="*60)
print("聚类分析详细报告")
print("="*60)

for cluster, info in cluster_report.items():
    print(f"\n簇{cluster} (占比: {info['percentage']:.1f}%):")
    print(f"  样本数量:  {info['size']}")
    print("  特征概况:")
    
    for feature, stats in info['characteristics'].items():
        print(f"    {feature}: 均值={stats['mean']:.2f}, 标准差={stats['std']:.2f}")

# 基于聚类结果进行业务解读和命名
def interpret_clusters(cluster_profile):
    interpretations = {}
    
    for cluster_id in cluster_profile.index:
        profile = cluster_profile.loc[cluster_id]
        
        # 基于特征进行业务逻辑判断
        high_income = profile['AnnualIncome'] > cluster_profile['AnnualIncome'].mean()
        high_spending = profile['SpendingScore'] > cluster_profile['SpendingScore'].mean()
        frequent_buyer = profile['TotalPurchases'] > cluster_profile['TotalPurchases'].mean()
        big_spender = profile['AvgTransactionValue'] > cluster_profile['AvgTransactionValue'].mean()
        
        # 根据业务逻辑给簇命名
        if high_income and high_spending and big_spender:
            name = "高价值客户"
            strategy = "重点维护，提供VIP服务和新品优先体验"
        elif high_income and not high_spending:
            name = "潜力客户" 
            strategy = "加强营销，提高消费意愿"
        elif not high_income and high_spending and frequent_buyer:
            name = "忠诚客户"
            strategy = "维持关系，提供忠诚度奖励"
        elif not high_income and not high_spending:
            name = "普通客户"
            strategy = "基础维护，尝试提升价值"
        else:
            name = "特殊客户"
            strategy = "进一步分析行为特征"
            
        interpretations[cluster_id] = {
            'name': name,
            'strategy': strategy,
            'profile': profile
        }
    
    return interpretations

# 生成业务解读
business_interpretation = interpret_clusters(cluster_profile)

print("\n" + "="*60)
print("业务解读与营销策略")
print("="*60)

for cluster_id, info in business_interpretation.items():
    print(f"\n簇{cluster_id}: {info['name']}")
    print(f"  营销策略:  {info['strategy']}")
    print(f"  客户特征:")
    print(f"    - 平均年收入: ${info['profile']['AnnualIncome']:,.0f}")
    print(f"    - 平均消费得分: {info['profile']['SpendingScore']:.1f}")
    print(f"    - 平均购买次数: {info['profile']['TotalPurchases']:.1f}")
    print(f"    - 平均交易价值:  ${info['profile']['AvgTransactionValue']:.1f}")

# 创建综合可视化仪表板
fig = plt.figure(figsize=(20, 12))

# 1. 聚类分布饼图
plt.subplot(2, 3, 1)
cluster_counts = df['Cluster'].value_counts()
plt.pie(cluster_counts.values, labels=[f'簇 {i}' for i in cluster_counts.index], 
        autopct='%1.1f%%', startangle=90)
plt.title('客户分群分布')

# 2. 收入vs消费得分散点图（按簇着色）
plt.subplot(2, 3, 2)
for cluster in sorted(df['Cluster'].unique()):
    cluster_data = df[df['Cluster'] == cluster]
    plt.scatter(cluster_data['AnnualIncome'], cluster_data['SpendingScore'], 
               label=f'´Ø {cluster}', alpha=0.7)
plt.xlabel('年收入')
plt.ylabel('消费得分')
plt.title('收入 vs 消费得分（按簇）')
plt.legend()

# 3. 特征重要性雷达图（简化版）
plt.subplot(2, 3, 3)
# 使用柱状图代替雷达图
feature_importance = cluster_profile[features_for_clustering].std().sort_values(ascending=False)
plt.barh(range(len(feature_importance)), feature_importance.values)
plt.yticks(range(len(feature_importance)), feature_importance.index)
plt.xlabel('标准差')
plt.title('特征区分度（标准差）')

# 4.簇特征热力图
plt.subplot(2, 3, 4)
sns.heatmap(cluster_profile[features_for_clustering], annot=True, cmap='YlOrRd', 
            fmt='.1f', linewidths=0.5)
plt.title('各簇特征均值热力图')

# 5.客户价值矩阵
plt.subplot(2, 3, 5)
for cluster in sorted(df['Cluster'].unique()):
    cluster_data = df[df['Cluster'] == cluster]
    plt.scatter(cluster_data['TotalPurchases'], cluster_data['AvgTransactionValue'], 
               label=f'´Ø {cluster}', alpha=0.7)
plt.xlabel('总购买次数')
plt.ylabel('平均交易价值')
plt.title('购买频率 vs 交易价值')
plt.legend()

plt.tight_layout()
plt.show()

# 保存聚类结果
df_with_clusters = df.copy()

# 添加业务标签
cluster_names = {cluster: info['name'] for cluster, info in business_interpretation.items()}
df_with_clusters['Cluster_Name'] = df_with_clusters['Cluster'].map(cluster_names)

# 保存到CSV文件
df_with_clusters.to_csv('customer_segmentation_results.csv', index=False)

# 生成总结报告
print("\n" + "="*60)
print("聚类分析总结报告")
print("="*60)
print(f"数据集大小: {len(df)}个客户")
print(f"最佳聚类数量: {best_k} 个细分群体")
print(f"整体轮廓系数: {silhouette_score(X_scaled, clusters):.4f}")
print(f"聚类效果: {'良好' if silhouette_score(X_scaled, clusters) > 0.5 else '一般'}")
print("\n各细分群体总结:")
for cluster_id in sorted(df['Cluster'].unique()):
    cluster_data = df[df['Cluster'] == cluster_id]
    print(f"  {business_interpretation[cluster_id]['name']}: {len(cluster_data)}人"
          f" ({len(cluster_data)/len(df)*100:.1f}%)")

print(f"\n结果已保存至:customer_segmentation_results.csv")






```
